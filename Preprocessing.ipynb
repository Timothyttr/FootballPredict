```python
all_files = glob.glob("*.csv")
df = pd.concat((pd.read_csv(f) for f in all_files))
df
```

```python
# all column names

pd.set_option('display.max_columns', None)
print("All available features:", df.columns.tolist())
```

```python
#Get the sum of missing values per column
nan_counts = df.isna().sum()
nan_counts = nan_counts[nan_counts > 0].sort_values(ascending=False)
print(nan_counts)
```

```python
# 1. Filter the dataframe to get only rows with at least one missing value
rows_with_nan = df[df.isna().any(axis=1)]

# 2. Print the count
print(f"Total rows with missing values: {len(rows_with_nan)} out of {len(df)}")

# 3. Show the first 5 and last 5 examples
# We transpose (.T) the head so you can actually read the columns
display(rows_with_nan.head().T) 
display(rows_with_nan.tail().T)
```

```python
target_cols = [
    'Div', 'Date', 'Time', 'HomeTeam', 'AwayTeam', 'Referee',
    
    # Full Time Results
    'FTHG', 'FTAG', 'FTR',
    
    # Half Time Results
    'HTHG', 'HTAG', 'HTR',
    
    # Match Stats
    'HS', 'AS',   # Shots
    'HST', 'AST', # Shots on Target
    'HC', 'AC',   # Corners
    'HF', 'AF',   # Fouls Committed
    'HFKC', 'AFKC', # Free Kicks Conceded
    'HO', 'AO',   # Offsides
    'HY', 'AY',   # Yellow Cards
    'HBP', 'ABP', # Booking Points
    
    # Betting Odds 
    'B365H', 'B365D', 'B365A',  # Primary
    'AvgH', 'AvgD', 'AvgA'      # Backup
]

available_cols = df.columns.intersection(target_cols)
df_clean = df[available_cols].copy()

if 'HBP' not in df_clean.columns and 'HY' in df_clean.columns and 'HR' in df_clean.columns:
    print("Generating Home Booking Points (HBP) from raw cards...")
    df_clean['HBP'] = (df_clean['HY'] * 10) + (df_clean['HR'] * 25)

if 'ABP' not in df_clean.columns and 'AY' in df_clean.columns and 'AR' in df_clean.columns:
    print("Generating Away Booking Points (ABP) from raw cards...")
    df_clean['ABP'] = (df_clean['AY'] * 10) + (df_clean['AR'] * 25)

if 'AvgH' in df_clean.columns:
    df_clean['B365H'] = df_clean['B365H'].fillna(df_clean['AvgH'])
    df_clean['B365D'] = df_clean['B365D'].fillna(df_clean['AvgD'])
    df_clean['B365A'] = df_clean['B365A'].fillna(df_clean['AvgA'])

df_clean = df_clean.dropna(subset=['FTHG', 'FTAG', 'HST', 'AST'])

df_clean['Date'] = pd.to_datetime(df_clean['Date'], dayfirst=True)
df_clean = df_clean.sort_values('Date')

print(f"Final Cleaned Data: {len(df_clean)} matches")
print("Columns kept:", list(df_clean.columns))
df_clean
```

```python
# 1. Check for missing values in the cleaned dataframe
nan_status = df_clean.isna().sum()

# 2. Filter to show ONLY columns with missing data
nan_status = nan_status[nan_status > 0].sort_values(ascending=False)

if len(nan_status) > 0:
    print("Columns with missing values:")
    print(nan_status)
else:
    print("No missing values! The dataset is 100% clean.")
```
